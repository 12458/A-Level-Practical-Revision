# Question 1
Use an OOP implementation to implement a fixed size Stack Abstract Data Type with the following operations:

- push(data: OBJECT), which inserts the OBJECT, data, at the top of the Stack.
- pop(), which removes and returns the OBJECT at the top of the Stack.

```python
class Stack:
    def __init__(self, size):
        self.size = size
        self.head = -1
        self.buffer = [None for _ in range(size)]
    
    def push(self, item):
        if self.isFull():
            return
        self.head += 1
        self.buffer[self.head] = item
    def pop(self):
        if self.isEmpty():
            return
        ret = self.buffer[self.head]
        self.buffer[self.head] = None
        self.head -= 1
        return ret
    def isEmpty(self):
        return self.head == -1
    def isFull(self):
        return self.head == self.size - 1
```

Implement a Circular Queue data structure with the following functionality:

- isEmpty(), which returns TRUE if the Queue is empty, or else returns FALSE.
- ifFull(), which returns TRUE is the Queue is full, or else returns FALSE.
- enqueue(data), which inserts the OBJECT, data, at the tail of the Queue.
- dequeue(), which removes and returns the OBJECT at the head of the Queue.

You should NOT use an object-oriented implementation.

```python
class CQ():
    def __init__(self, k):
        self.k = k
        self.queue = [None] * k
        self.head = self.tail = -1
    # Insert an element into the circular queue
    def enqueue(self, data):
        if self.full():
            print("The circular queue is full\n")
        elif (self.head == -1):
            self.head = 0
            self.tail = 0
            self.queue[self.tail] = data
        else:
            self.tail = (self.tail + 1) % self.k
            self.queue[self.tail] = data
    
    def full(self):
        return ((self.tail + 1) % self.k == self.head)
    
    def empty(self):
        return self.head == -1
    # Delete an element from the circular queue
    def dequeue(self):
        if self.empty():
            return
        elif (self.head == self.tail):
            temp = self.queue[self.head]
            self.head = -1
            self.tail = -1
            return temp
        else:
            temp = self.queue[self.head]
            self.head = (self.head + 1) % self.k
            return temp
```

# Question 4
Implement a Singly-linked Linked-list (SLLL) data structure with the following functionality:
- insertFront(data), which inserts the OBJECT, data, at the front the Linked-list.
- insertBack(data), which inserts the OBJECT, data, at the back the Linked-list.
- sort(), which sorts the contents of the Linked-list.
- exists(data), which returns TRUE if the OBJECT, data, exists within the Linked-list.
- delete(data), which removes the OBJECT, data from the Linked-list, and either returns TRUE if successful, or else returns False.

You should implement this data structure using object-oriented programming.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
class LinkedList:
    def __init__(self):
        self.first = None
        
    def sort(self):
        if self.first != None:
            tmp = []
            cur = self.first
            while cur != None:
                tmp.append(cur.data)
                cur = cur.next
            tmp.sort()
            for i in tmp:
                self.insertBack(i)
     
    def insertFront(self, data):
        new_node = Node(data)
        if self.first == None:
            self.first = new_node
            return
        # Order is important
        new_node.next = self.first
        self.first = new_node
        return
    
    def insertBack(self, data):
        new_node = Node(data)
        if self.first == None:
            self.first = new_node
            return
        cur = self.first
        while cur != None:
            cur = cur.next
        cur.next = new_node
        return
    def exists(self, data):
        cur = self.first
        while cur != None:
            if cur.data == data:
                return True
            cur = cur.next
        return False
    def delete(self, data):
        # 1st case
        if self.first == None:
            return False
        # 2nd case
        if self.first.data == data:
            self.first = self.first.next
            return True
        # 3rd case
        prev = self.first
        cur = self.first.next
        delete_occured = False
        while cur != None:
            if cur.data == data:
                prev.next = cur.next
                delete_occured = True
            prev = cur
            cur = cur.next
        if delete_occured:
            return True
        return False
```

# Question 5

- insert(data), which inserts data into the SLLL such that the SLLL is ordered in ascending value of the data item.
- delete(data)

You should implement this data structure using object-oriented programming.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
class LinkedList:
    def __init__(self):
        self.first = None
  
    def insert(self, data):
        new_node = Node(data)
        if self.first == None:
            self.first = new_node
            return
        cur = self.first
        while cur != None:
            if cur.data <= data:
                cur = cur.next
                continue
            else:
                break
        tmp = cur.next
        new_node.next = tmp
        cur.next = new_node
        return
    def delete(self, data):
        # 1st case
        if self.first == None:
            return False
        # 2nd case
        if self.first.data == data:
            self.first = self.first.next
            return True
        # 3rd case
        prev = self.first
        cur = self.first.next
        delete_occured = False
        while cur != None:
            if cur.data == data:
                prev.next = cur.next
                delete_occured = True
            prev = cur
            cur = cur.next
        if delete_occured:
            return True
        return False
```
