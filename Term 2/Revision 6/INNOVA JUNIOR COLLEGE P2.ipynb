````python
# Task 2.2
class HashTable:
    def __init__(self, n):
        self._array = [None for _ in range(n)]
        self._size = n

    def hash(self, s):
        return s % self._size

    def insert(self, data):
        index = self._exists(data, True)
        if index == -1:
            return False
        else:
            self._array[index] = data

    # Task 2.3
    def _exists(self, target, insert_flg=False):
        index = self.hash(int(target))
        if insert_flg == True:
            target = None
        if self._array[index] == target:
            return index
        else:
            i = (index + 1) % self._size
            while i != index:
                if self._array[i] == target:
                    return i
                else:
                    i = (i + 1) % self._size
            return -1

    def exists(self, target):
        index = self._exists(target)
        return (True, index) if index != -1 else (False, -1)

    def __str__(self):
        return str(self._array)


X = 12
ht = HashTable(X)
with open("KEYS2.TXT", "r") as f:
    for line in f:
        ht.insert(int(line.strip()))
print(ht)
````

````python
# Task3.1
class Node:
    def __init__(self, data):
        self._data = data  # STRING
        self._left = -1  # INT
        self._right = -1  # INT

    def setData(self, s):
        self._data = s

    def getData(self):
        return self._data

    def setLeftPtr(self, x):
        self._left = x

    def getLeftPtr(self):
        return self._left

    def setRightPtr(self, y):
        self._right = y

    def getRightPtr(self):
        return self._right

    def __str__(self):
        return f"{self._data}:{self._left}:{self._right}"


class BST:
    def __init__(self, size):
        self.size = size
        self.count = 0
        self.thisTree = [Node(None) for _ in range(size)]
        for i in range(self.size - 1):
            self.thisTree[i].setLeftPtr(i + 1)
        self.root = -1  # root of BST
        self.free = 0  # free node linked list

    def __str__(self):
        ret = f"{self.root},{self.free}\t"
        for i in range(self.size):
            ret += f"{self.thisTree[i]},"
        return ret[:-1]

    def add(self, data):
        # check if array is full
        if self.free == -1:
            return False
        # get the free node
        free_node_i = self.free
        # update the free node list
        self.free = self.thisTree[self.free].getLeftPtr()

        # update the free node: transition from a linked list node to a BST node
        self.thisTree[free_node_i].setData(data)
        self.thisTree[free_node_i].setLeftPtr(-1)

        # empty BST
        if self.root == -1:
            self.root = free_node_i
            return True

        cur_i = self.root
        while True:
            if data < self.thisTree[cur_i].getData():
                if self.thisTree[cur_i].getLeftPtr() == -1:
                    self.thisTree[cur_i].setLeftPtr(free_node_i)
                    return True
                else:
                    cur_i = self.thisTree[cur_i].getLeftPtr()
            else:
                if self.thisTree[cur_i].getRightPtr() == -1:
                    self.thisTree[cur_i].setRightPtr(free_node_i)
                    return True
                else:
                    cur_i = self.thisTree[cur_i].getRightPtr()

    # Task 3.3
    def postOrderTraversal(self):
        def _recur(node_i):
            if node_i == -1:
                return
            ## L, R, N

            _recur(self.thisTree[node_i].getLeftPtr())
            _recur(self.thisTree[node_i].getRightPtr())
            print(self.thisTree[node_i].getData())

        _recur(self.root)
````

````python
# Task 4.1
ten = "X"
strike = ten


def pins(throws):
    if throws == ten:
        return 10
    else:
        return int(throws)


def bowling_score(throws):
    def bowling_score_helper(throws, frame_no):
        # Frame 10 with no bonus
        if frame_no == 10 and len(throws) == 2:
            return pins(throws[0]) + pins(throws[1])

        # Frame 10 with bonus
        if frame_no == 10 and len(throws) == 3:
            return pins(throws[0]) + pins(throws[1]) + pins(throws[2])

        # A Strike
        if throws[0] == strike:
            frame_score = pins(throws[1]) + pins(throws[2])
            return frame_score + bowling_score_helper(throws[1:], frame_no + 1)

        frame_score = pins(throws[0]) + pins(throws[1])

        # A Spare
        if frame_score == 10:
            return 10 + pins(throws[2]) + bowling_score_helper(throws[2:], frame_no + 1)

        # Frame with no bonus
        return frame_score + bowling_score_helper(throws[2:], frame_no + 1)

    return bowling_score_helper(throws, 1)
````
