```python
# Task 2.1
import json
with open('RESTAURANTS.JSON') as j:
    resturants_raw = json.load(j)
seen_before = []
resturants = []
for resturant in resturants_raw:
    if resturant['name'] not in seen_before:
        seen_before.append(resturant['name'])
        if resturant['rating'] == 'Not yet rated':
            resturant['rating'] = 0
        resturants.append(resturant)
```

```python
resturants
```

```python
# Task 2.2
# Merge Sort
def merge(A, B):
    ret = []
    while len(A) and len(B):
        if A[0]['rating'] > B[0]['rating']:
            ret.append(A[0])
            A.pop(0)
        else:
            ret.append(B[0])
            B.pop(0)
    if len(A) == 0:
        ret += B
    else:
        ret += A
    return ret


def mergeSort(L):
    if len(L) <= 1:
        return L
    mid = len(L) // 2
    return merge(mergeSort(L[:mid]), mergeSort(L[mid:]))
```

```python
print(mergeSort(resturants)[:10])
```

```python
# Task 2.4
import pymongo
c = pymongo.MongoClient('localhost', 27017)

c['Entertainment']['Resturants'].insert_many(mergeSort(resturants))

c.close()
```
