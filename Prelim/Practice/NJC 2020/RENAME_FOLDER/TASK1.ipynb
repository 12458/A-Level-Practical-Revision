```python
# Task 1.1
def getSecret():
    words = {}
    words_raw = []
    with open('WORDS.txt') as f:
        for line in f:
            words_raw.append(line.strip())

    for word in words_raw:
        if len(word) in words.keys():
            words[len(word)].append(word.lower())
        else:
            words[len(word)] = [word]
    shortest_word = min(words.keys())
    longest_word = max(words.keys())
    len_guess = int(input(f'Choose the length of word to guess: [{shortest_word} to {longest_word}] > '))
    import random
    return random.sample(words[len_guess], 1)[0]
```

```python
# Task 1.2

def mask(word:str, guessed_char):
    return ''.join([i if i in guessed_char else '.' for i in word])

def game():
    word = getSecret()
    #print(word)
    guesses = len(word) * 2
    guessed_char = []
    print(f'You have {guesses} attempts. Good Luck!')
    print(mask(word, guessed_char))
    for attempt in range(guesses, 0, -1):
        guess = input(f'[{attempt}] Guess letter: ').lower()
        if guess in word:
            guessed_char.append(guess)
        print(mask(word, guessed_char))
        if '.' not in mask(word, guessed_char):
            print('WIN')
            return
    print(f'LOSE ... The word is {word}')
    return
```
