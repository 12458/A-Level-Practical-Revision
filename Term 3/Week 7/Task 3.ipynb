```python
# Task 3.2
# Task 3.1
import csv
with open('surface_air_temperature.csv') as f:
    c = csv.reader(f)
    next(c)
    surface_air_temperature = list(c)
    
for i in range(len(surface_air_temperature)):
    surface_air_temperature[i][0] = surface_air_temperature[i][0].split('-')

max_temp = {}
min_temp = {}

for entry in surface_air_temperature:
    if entry[0][0] in max_temp.keys():
        max_temp[int(entry[0][0])] = max(float(max_temp[int(entry[0][0])]), float(entry[2]))
    else:
        max_temp[int(entry[0][0])] = float(entry[2])
    if entry[0][0] in min_temp.keys():
        min_temp[int(entry[0][0])] = min(float(min_temp[int(entry[0][0])]), float(entry[1]))
    else:
        min_temp[int(entry[0][0])] = float(entry[1])

for year in min_temp.keys():
    print(f"Year: {year} | Min Temp: {min_temp[year]} (Month: {[i[1] for i in surface_air_temperature if i[0][0] == str(year)].index(str(min_temp[year]).replace('.0',''))}) | Max Temp: {max_temp[year]} (Month: {[i[2] for i in surface_air_temperature if i[0][0] == str(year)].index(str(max_temp[year]).replace('.0',''))})")
```

```python
avg_min = {}
avg_max = {}

for month in ['01','02','03','04','05','06','07','08','09','10','11','12']:
    tmp = [float(i[1]) for i in surface_air_temperature if i[0][1] == month]
    avg_min[month] = sum(tmp) / len(tmp)

for month in ['01','02','03','04','05','06','07','08','09','10','11','12']:
    tmp = [float(i[2]) for i in surface_air_temperature if i[0][1] == month]
    avg_max[month] = sum(tmp) / len(tmp)

    
```
