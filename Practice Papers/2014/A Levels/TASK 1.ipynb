```python
menu = \
"""
++++++++++++++++++++++++++++++++++++++++++++++
1. An exact match
2. A match at the beginning of the term text
3. A match anywhere within the term text
++++++++++++++++++++++++++++++++++++++++++++++
"""

with open('JARGON.TXT') as f:
    jargon = f.read().strip().split('\n')

while True:
    print(menu)
    choice = int(input('CHOICE? > '))
    if choice == 1:
        term = input('ENTER TERM > ')
        if term in jargon:
            print(term)
        print(f'There were {jargon.count(term)} matching term(s)')
    elif choice == 2:
        term = input('ENTER TERM > ')
        matches = [i for i in jargon if i[:len(term)] == term]
        for match in matches:
            print(match)
        print(f'There were {len(matches)} matching term(s)')
    else:
        term = input('ENTER TERM > ')
        matches = [i for i in jargon if (term in i)]
        for match in matches:
            print(match)
        print(f'There were {len(matches)} matching term(s)')
    
```
