````python
n = int(input('NUMBER OF PEOPLE TO ENTER > '))
assert n <= 10
steps = []
for i in range(n):
    detail = input('ENTER DETAILS [name,steps]: ').strip().split(',')
    detail[1] = int(detail[1])
    steps.append(detail)
steps = sorted(steps, key=lambda x:x[1])

with open('STAR.TXT') as f:
    last_star = f.read().strip().split('\n,')
    
print(f"Last week, {last_star[0]} was 'Star of the week' with {last_star[1]} steps")
print(f"This week, {steps[-1][0]} is 'Star of the week' with {steps[-1][1]} steps")

with open('STAR.TXT', 'w') as f:
    f.write(f'{steps[-1][0]}\n,{steps[-1][1]}')
````
