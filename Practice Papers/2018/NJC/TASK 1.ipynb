````python
# Task 1.2
def bubble_sort_means(data:list):
    for i in range(len(data)-1):
        #print(i)
        swap = False
        for j in range(len(data)-1-i):
            if sum(data[j][1:])/3 < sum(data[j+1][1:])/3:
                swap = True
                #print(f'{sum(data[j][1:])/3} < {sum(data[j+1][1:])/3}')
                data[j], data[j+1] = data[j+1] , data[j]
        if swap == False:
            break
    return data
````

````python
# Task 1.4
def insertion_sort_dates(A):
    for i in range(1, len(A)):
        key = A[i]
        j = i - 1
        while j >= 0 and A[j][0] > key[0]:
            A[j + 1] = A[j]
            j -= 1
        A[j + 1] = key
    return A
````
