````python
class ToDo:
    def __init__(self, c:str, d:str):
        self._category = c
        self._description = d
        
    def get_category(self):
        return self._category
    
    def get_description(self):
        return self._description
    
    def set_description(self, d):
        self._description = d
        
    def set_category(self, c):
        self._category = c
    
    def summary(self):
        return str(self._category) + ',' + str(self._description)  
    
    def compare_with(self, td):
        if self._category < td.get_category():
            return -1
        elif self._category > td.get_category():
            return 1
        else:
            if self._description == td.get_description():
                return 0
            elif self._description < td.get_description():
                return -1
            else:
                return 1
    
    def __eq__(self, x):
        # print(f"Comparing {str(x)} and {str(self)} | {x._category == self._category and x._description == self._description}")
        try:
            return x._category == self._category and x._description == self._description
        except:
            return False
    
    def __repr__(self):
        return str(self._category) + ',' + str(self._description)
````

````python
list_todo = []
# No need to read file
with open('2019_TASK3_2.txt') as f:
    for l in f:
            list_todo.append(eval(l.strip()))
def insertion_sort(A):
    for i in range(len(A)):
        j = i
        while j > 0 and A[j - 1].compare_with(A[j]) == 1:
            A[j], A[j - 1] = A[j - 1], A[j]
            j = j - 1

insertion_sort(list_todo)
        
[print(i.summary()) for i in list_todo]
pass
````

````python
class DatedToDo(ToDo):
    def __init__(self, dt, c, d):
        self._due_date = dt
        super().__init__(c, d)
    
    def get_due_date(self):
        return self._due_date
    
    def set_due_date(self, dt):
        self._due_date = dt
    
    def summary(self):
        return str(self._category) + ',' + str(self._description) + ',' + str(self._due_date)
        
    def compare_with(self, td):
        try:
            if self._due_date < td.get_due_date():
                return -1
            elif self._due_date > td.get_due_date():
                return 1
            else:
                super().compare_with(td)
        except:
            super().compare_with(td)
            
    def __repr__(self):
        return str(self._category) + ',' + str(self._description) + ',' + str(self._due_date)
    
    def __eq__(self, x):
        # print(f"Comparing {str(x)} and {str(self)} | {x._category == self._category and x._description == self._description and x._due_date == self._due_date}")
        try:
            return x._category == self._category and x._description == self._description and x._due_date == self._due_date
        except:
            return False

dtd_l = []
td_l = []

with open('2019_TASK3_3.txt') as f:
    for l in f:
        if isinstance(eval(l.strip()), DatedToDo): # type(obj) will give you the exact object
            # isinstance(x, y) will ret True if x inherits from y
            dtd_l.append(eval(l.strip()))
        elif isinstance(eval(l.strip()), ToDo):
            td_l.append(eval(l.strip()))

insertion_sort(dtd_l)
insertion_sort(td_l)

all_td = dtd_l + td_l
[print(i.summary()) for i in all_td]
pass
````
