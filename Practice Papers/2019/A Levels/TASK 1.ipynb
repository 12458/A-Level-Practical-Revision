FORMAT
---
`YYYY-MM-DD\tHH:MM\tTIDE\tHEIGHT\n`

`TIDES.TXT`
---
```
2017-08-01	02:40	LOW	1.2
2017-08-01	09:02	HIGH	5.9
2017-08-01	14:47	LOW	1.6
2017-08-01	21:31	HIGH	5.8
2017-08-02	03:42	LOW	1.4
2017-08-02	10:00	HIGH	5.8
...
```

````python
import csv
with open('TIDES.TXT') as f:
    c = csv.reader(f, delimiter='\t')
    tide_data = list(c)

low_tide = [record for record in tide_data if record[2] == 'LOW']
high_tide = [record for record in tide_data if record[2] == 'HIGH']

low_tide_sorted = sorted(low_tide, key=lambda x: x[3])
high_tide_sorted = sorted(high_tide, key=lambda x: x[3])

print(low_tide_sorted[0][3])
print(high_tide_sorted[-1][3])
````

````python
tide_data
````

````python
# Assumption: Difference between tide is the absolute difference

low_to_high = []
for i in range(0,len(tide_data),2):
    low_to_high.append([tide_data[i], tide_data[i+1], float(tide_data[i+1][3]) - float(tide_data[i][3])])
    
high_to_low = []
for i in range(1,len(tide_data)-1,2):
    high_to_low.append([tide_data[i], tide_data[i+1], float(tide_data[i][3]) - float(tide_data[i+1][3])])
````

````python
sorted(low_to_high, key=lambda x: x[2])
````

````python
sorted(high_to_low, key=lambda x: x[2])
````

````python
tidal_ranges = low_to_high + high_to_low
sorted_tidal_range = sorted(tidal_ranges, key=lambda x: x[2])

print(f'LOWEST TIDAL RANGE: {sorted_tidal_range[0][2]}, DATE: {sorted_tidal_range[0][1][0]}')
print(f'LARGEST TIDAL RANGE: {sorted_tidal_range[-1][2]}, DATE: {sorted_tidal_range[-1][1][0]}')
````
