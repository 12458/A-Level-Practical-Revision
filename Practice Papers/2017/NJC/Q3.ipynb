```python
# Task 3.1
class BSTNode:
    pass
    def __init__(self,data):
        pass
        self._data = data
        self._left = None
        self._right = None
        
    def get_data(self):
        return self._data
    
    def set_data(self,data):
        self._data = data
        
    def get_left(self):
        return self._left
    
    def set_left(self,data):
        self._left = data
        
    def get_right(self):
        return self._right
    
    def set_right(self,data):
        self._right = data
    
    def print(self):
        print(self._data)
        
class LLNode(BSTNode):
    def __init__(self,data):
        super().__init__(data)
        self._next = None
    
    def get_next(self):
        return self._next
    
    def set_next(self,data):
        self._next = data
    
    def insert(self,data):
        new_node = BSTNode(data)
        cur = self
        while True:
            if data < cur.get_data():
                # Insert left
                if cur.get_left() == None:
                    cur.set_left(new_node)
                    return
                else:
                    cur = cur.get_left()
            else:
                # Insert right
                if cur.get_right() == None:
                    cur.set_right(new_node)
                    return
                else:
                    cur = cur.get_right()
                    
    def bst_print(self):
        def _inOrder(node):
            if node == None:
                return
            _inOrder(node.get_left())
            node.print()
            _inOrder(node.get_right())
        _inOrder(self)
        print()
        
    def exists(self, data):
        cur = self

        def _exists(cur, data):
            if cur.get_data() == data:
                return True
            if data < cur.get_data():
                if cur.get_left() == None:
                    return False
                return _exists(cur.get_left(), data)
            else:
                if cur.get_right() == None:
                    return False
                return _exists(cur.get_right(), data)
        return _exists(cur, data)
```

```python
# Task 3.2
class HybridStructure:
    def __init__(self):
        self._first = None
    
    def get_first(self):
        return self._first
    
    def set_first(self,data):
        self._first = data
    
    def inLL(self,data):
        tens = int(str(data).zfill(2)[0])
        cur = self._first
        while cur != None:
            cur_tens = int(str(cur.get_data()).zfill(2)[0])
            if cur_tens == tens:
                return True
            cur = cur.get_next()
        return False
    
    def contains(self,data):
        if not self.inLL(data):
            return False
        tens = int(str(data).zfill(2)[0])
        cur = self._first
        while cur != None:
            cur_tens = int(str(cur.get_data()).zfill(2)[0])
            if cur_tens == tens:
                return cur.exists(data)
            cur = cur.get_next()
        return False
    
    def insert(self,data):
        if self.inLL(data):
            tens = int(str(data).zfill(2)[0])
            cur = self._first
            while cur != None:
                cur_tens = int(str(cur.get_data()).zfill(2)[0])
                if cur_tens == tens:
                    cur.insert(data)
                    return
                cur = cur.get_next()
        else:
            new_node = LLNode(data)
            if self._first == None:
                self._first = new_node
                return
            new_node.set_next(self._first)
            self._first = new_node
            return
       
    
    def print(self):
        cur = self._first
        while cur != None:
            cur.bst_print()
            cur = cur.get_next()
```
