```python
import csv
with open('data-gov-sg-dataset-listing.csv') as f:
    c = csv.DictReader(f, fieldnames = ['_id','organisation','dataset_id','dataset_name','date_created','last_updated','description','frequency','coverage_start','coverage_end','resource_id','resource_name','resource_description','resource_format'])
    next(c)
    data = list(c)
    for i in range(len(data)):
        data[i]['_id'] = int(data[i]['_id'])
```

```python
data
```

```python
import pymongo
c = pymongo.MongoClient('localhost', 27017)
```

```python
coll = c['GovTech']['Datasets']
```

```python
coll.insert_many(data)
```
