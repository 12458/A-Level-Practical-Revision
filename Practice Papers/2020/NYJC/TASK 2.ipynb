````python
import csv
with open('story.csv') as f:
    c = csv.reader(f)
    next(c)
    story = list(c)
    # with implictly closes the file object when exit
sort_by = input('SORT BY [B(ook), S(ubject), A(uthor name) or P(ublished)]').strip()
assert sort_by in ['A', 'S', 'B', 'P']
translate = {
        'B': 0,
        'S': 1,
        'A': 2,
        'P': 3,
    }
ArrayData = MergeSort(story, translate[sort_by])
print(ArrayData)
````

````python
# Merge Sort
def merge(A, B, sortby):
    ret = []
    while len(A) and len(B):
        if A[0][sortby] < B[0][sortby]:
            ret.append(A[0])
            A.pop(0)
        else:
            ret.append(B[0])
            B.pop(0)
    if len(A) == 0:
        ret += B
    else:
        ret += A
    return ret


def MergeSort(Array, sortby):
    if len(Array) <= 1:
        return Array
    mid = len(Array) // 2
    return merge(MergeSort(Array[:mid], sortby), MergeSort(Array[mid:], sortby), sortby)
````

````python
def quick_sort(A):
    if len(A) < 2:
        return A
    pivot = A[0]
    left = []
    right = []
    for i in A[1:]:
        if i[0] > pivot[0]:
            left.append(i)
        else:
            right.append(i)
    return quick_sort(left) + [pivot] + quick_sort(right)
````
