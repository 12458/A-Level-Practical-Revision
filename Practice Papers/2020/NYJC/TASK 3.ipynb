```python
class SongPlayList:
    def __init__(self):
        self.head = None
    
    def insert(self, title):
        new_node = Node(title)
        if self.head == None:
            self.head = new_node
            return True
        
        # update ptrs
        new_node.next = self.head
        self.head.prev = new_node
        self.head = new_node      
        return True
    
    def insertafter(self, searchtitle, newtitle):
        new_node = Node(newtitle)
        if self.head == None:
            # DLL is empty; insert at head
            self.head = new_node
            return True
                
        # traverse and find searchtitle
        found = False
        count = 0
        cur = self.head
        while cur != None:
            if cur.title == searchtitle:
                found = True
                break
            count += 1
            cur = cur.next
        if found == False:
            return False
        
        # update ptrs
        next_node = cur.next
        prev_node = cur
        
        new_node.next = next_node
        new_node.prev = prev_node
        
        if next_node != None:
            next_node.prev = new_node
            
        prev_node.next = new_node
        
        return True
    
    def insert_in_sorted_order(self, title):
        new_node = Node(title)
        if self.head == None:
            # DLL is empty; insert at head
            self.head = new_node
            return True
        
        # insert front
        if title < self.head.title:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node
            return True
                
        # traverse and find correct location
        cur = self.head
        while cur != None and cur.next != None:
            if cur.title > cur.next.title:
                break
            cur = cur.next      
        
        # update ptrs
        next_node = cur.next
        prev_node = cur
        
        new_node.next = next_node
        new_node.prev = prev_node
        
        if next_node != None:
            next_node.prev = new_node
            
        prev_node.next = new_node
        
        return True
    
    def insertionSort(self):
        new = SongPlayList()
        cur = self.head
        while cur != None:
            new.insert_in_sorted_order(cur.title)
            cur = cur.next
        new.display()
        self.head = new.head
        return
    
    def display(self):
        # Check if empty
        if self.head == None:
            print('EMPTY PLAYLIST')
        else:
            ret = ''
            cur = self.head
            while cur != None:
                ret += str(cur) + ','
                cur = cur.next
            print(ret[:-1])

class Node:
    def __init__(self, title):
        self.title = title
        self.next = None
        self.prev = None
    
    def __repr__(self):
        return f'{self.title}'
```
