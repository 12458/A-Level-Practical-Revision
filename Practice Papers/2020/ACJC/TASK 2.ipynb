```python
class Game:
    def __init__(self, n):
        self.size = n
        self.board = [[(i,n+1-j) for i in range(1,n+1)] for j in range(1, n+1)]
    
    def validate_knight(self, cur_pos, new_pos, mode):
        if mode == 1:
            if not (1 <= new_pos[0] <= self.size and 1 <= new_pos[1] <= self.size):
                # exceed condition
                return False
        
        # print(cur_pos, new_pos)
        valid_moves = []
        up = lambda pos: (pos[0],pos[1]+1)
        down = lambda pos: (pos[0],pos[1]-1)
        left = lambda pos: (pos[0]-1,pos[1])
        right = lambda pos: (pos[0]+1,pos[1])
        # bounds are 1 --> n
        # case 1
        # L, L, U/D
        valid_moves.append(up(left(left(cur_pos))))
        valid_moves.append(down(left(left(cur_pos))))
        # case 2
        # R, R, U/D
        valid_moves.append(up(right(right(cur_pos))))
        valid_moves.append(down(right(right(cur_pos))))
        
        # case 3
        # D, D, L/R
        valid_moves.append(left(down(down(cur_pos))))
        valid_moves.append(right(down(down(cur_pos))))
        
        # case 4
        # U, U, L/R
        valid_moves.append(left(up(up(cur_pos))))
        valid_moves.append(right(up(up(cur_pos))))
        
        # print('valid', valid_moves)
        if mode == 1:
            return new_pos in valid_moves
        else:
            return [i for i in valid_moves if 1 <= i[0] <= self.size and 1 <= i[1] <= self.size]
    
    def avaliable(self, square, lis):
        potential_valid_moves = self.validate_knight(square, None, 0)
        potential_valid_moves = [i for i in potential_valid_moves if i not in lis]
        return sorted(potential_valid_moves)
    
    def check_tour(self, tour_path):
        # A knightâ€™s tour is a sequence of moves that a chess knight makes on a chessboard,  so  that  it 
        # visits every square of the chessboard __exactly__ once.
        
        # check if all items are unique
        
        unique_coordinates = list(set(tour_path))
        check_coords_unique = sum(
            [True if tour_path.count(i) == 1 else False for i in unique_coordinates]
        ) == len(unique_coordinates)
        
        if check_coords_unique == False:
            # print('coords not unique')
            return False
        
        validate_moves = []
        cur = tour_path[0]
        for next_pos in tour_path[1:]:
            valid = self.validate_knight(cur, next_pos, 1)
            # print(f'{cur}-->{next_pos} : {valid}')
            validate_moves.append(valid)
            cur = next_pos
        
        if False in validate_moves:
            # print('invalid moves')
            # print(validate_moves)
            return False
        
        return True
    
    def find_smallest_avaliable_square(self):
        cur = (1,1)
        lis = [(1,1)]
        while True:
            potential = sorted([(i, len(self.avaliable(i, lis+[i]))) for i in self.avaliable(cur, lis)],
                               key=lambda x:x[1])
            print(f'lis: {lis} | potential: {potential}')
            if len(potential) == 0:
                break
            # we select smallest 
            if self.check_tour(lis + [potential[0][0]]):
                lis.append(potential[0][0])
                cur = potential[0][0]
            else:
                # terminate
                break
        return lis
```
