```python
class TicTacTomek:
    def __init__(self):
        # init 4x4 grid
        import random
        self.grid = [['.' for _ in range(4)] for _ in range(4)]
        self.tomek = (random.randint(0,3), random.randint(0,3)) # tomek random location
        self.grid[self.tomek[1]][self.tomek[0]] = 'T'
    
    def displayBoard(self):
        print('\n'.join([''.join(i) for i in self.grid]))
    
    def getPlayerMove(self, player:str):
        while True:
            try:
                move = input(f'Player {player}: Enter Move [x,y]> ').strip().split(',')
                assert move[0].isdigit()
                assert move[1].isdigit()
                move = (int(move[0]), int(move[1]))
                assert self.grid[move[1]][move[0]] == '.'
                break
            except:
                print('Error in input!')
                continue
            
        if player == 'O':
            self.grid[move[1]][move[0]] = 'O'
        else:
            self.grid[move[1]][move[0]] = 'X'
        
        return True
    
    def checkWin(self, player):
        def chk_col(player):
            for i in range(4):
                tmp = (self.grid[0][i] == player or self.grid[0][i] == 'T')
                for j in range(4):
                    tmp = tmp and (self.grid[j][i] == player or self.grid[j][i] == 'T')
                if tmp == True:
                    return True
                else:
                    continue
            return False
        
        def chk_row(player):
            for i in range(4):
                tmp = (self.grid[i][0] == player or self.grid[i][0] == 'T')
                for j in range(4):
                    tmp = tmp and (self.grid[i][j] == player or self.grid[i][j] == 'T')
                if tmp == True:
                    return True
                else:
                    continue
            return False
                    
        
        def chk_diag(player):
            left_diag = (self.grid[0][0] == player or self.grid[0][0] == 'T')
            for i in range(1,4):
                left_diag = left_diag and (self.grid[i][i] == player or self.grid[i][i] == 'T')
            
            right_diag = (self.grid[0][3] == player or self.grid[0][3] == 'T')
            for i in range(1,4):
                right_diag = left_diag and (self.grid[i][3-i] == player or self.grid[i][3-i] == 'T')
            
            return left_diag or right_diag 
        
        return chk_row(player) or chk_col(player) or chk_diag(player)   
```

```python
# Task 4.5
def main():
    ttt = TicTacTomek()
    while True:
        for player in ['X', 'O']:
            ttt.displayBoard()
            ttt.getPlayerMove(player)
            if ttt.checkWin(player):
                print(f'{player} has won')
                ttt.displayBoard()
                return
            elif sum(['.' in i for i in ttt.grid]) == 0:
                print('Draw')
                return
```

```python
main()
```
