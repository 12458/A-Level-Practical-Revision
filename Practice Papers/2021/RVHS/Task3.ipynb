```python
# Task 3.1
# Write program code to read the csv file "health_facilities.csv" and insert all information in the file as documents
# into a NoSQL MongoDB database called "Health" with one collection called "facilities".
# The "_id" of the documents in the database should start from 1, 2, 3 and 4 etc. 
# The correct data type of each field is expected to be inserted into the database.

from pymongo import MongoClient
import csv

with open('health_facilities.csv') as f:
    c = csv.DictReader(f, fieldnames=['year','institution_type','facility_type_a','public_private','no_of_facilities','no_beds'])
    next(c)
    c = list(c)
    for i in range(len(c)):
        c[i]['year'] = int(c[i]['year'])
        c[i]['no_of_facilities'] = int(c[i]['no_of_facilities'])
        c[i]['no_beds'] = int(c[i]['no_beds'])

client = MongoClient('localhost',27017)
client['Health']['facilities'].insert_many(c)
```

```python
task3_3 = list(client['Health']['facilities'].find({'public_private':'Not-for-Profit', 'no_of_facilities': 0}, {'_id':1}))
```

```python
def bubble_sort(A):
    for i in range(len(A)-1):
        for j in range(len(A)-i-1):
            if A[j]['no_beds'] > A[j+1]['no_beds']:
                # swap
                A[j], A[j+1] = A[j+1], A[j]
    return A
```
