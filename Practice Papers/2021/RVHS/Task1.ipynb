```python
# Task 1.1
# Write a function task1_1(name_A, name_B) where name_A and name_B are strings 
# which consists of alphabet letters and spaces. The function will return True if 
# A) The alphabet letters used in string name_A and name_B are the same and
# B) the spaces in string name_A and name_B are at the same locations.
# For example,
# >>> match_names("Abcde", "Deabc")
# True
# >>> match_names("Abcde Fgh I", "Ihgfe Dcb A")
# True
# >>> match_names("Abcd Efgh I", "Ihgfe Dcb A")
# False
# >>> match_names("Abcde Fzh I", "Ihgfe Dcb A")
# False
def task1_1(name_A, name_B):
    name_A = name_A.upper()
    name_B = name_B.upper()
    
    def findall(p, s):
        '''Yields all the positions of
        the pattern p in the string s.'''
        i = s.find(p)
        while i != -1:
            yield i
            i = s.find(p, i+1)
        
    # locate spaces
    space_A = list(findall(' ', name_A))
    space_B = list(findall(' ', name_B))
    unique_char_A = list(set(name_A))
    unique_char_B = list(set(name_B))
    
    if space_A != space_B:
        return False
    
    if len(name_A) == len(name_B):
        for char in unique_char_A:
            # print(char)
            # print(name_A.count(char), name_B.count(char))
            if name_A.count(char) != name_B.count(char):
                return False
        return True
    else:
        return False

def quick_test():
    print(task1_1("Abcde", "Deabc")) #True
    print(task1_1("Abcde Fgh I", "Ihgfe Dcb A")) #True
    print(task1_1("Abcd Efgh I", "Ihgfe Dcb A")) #False
    print(task1_1("Abcde Fzh I", "Ihgfe Dcb A")) #False
quick_test()
```

```python
def task1_2():
    with open('namelist_A.txt') as f:
        nameA = f.read().strip().split('\n')
    
    with open('namelist_B.txt') as f:
        nameB = f.read().strip().split('\n')
        
    print(f'{"From namelist_A.txt":<40}{"From namelist_B.txt":<40}')
    
    for record in nameA:
        find = [i for i in nameB if task1_1(record, i) == True]
        print(f'{record:<40}{find[0] if len(find) > 0 else "*******NO MATCH*******":<40}')
    
task1_2()
```
