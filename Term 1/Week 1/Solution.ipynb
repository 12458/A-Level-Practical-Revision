# Q3

Please implement a Binary Search function.

Your implementation should take in 2 parameters:
- an array of sorted objects (in ascending order), and
- a target object.

Your implementation should also return the index of the first instance of the target object within the given array, or else, if not found, returns -1.

```python
def binarySearch(arr, target, lb, ub):
    mid_idx = (lb + ub)//2
    if target == arr[mid_idx]:
        return mid_idx
    else if target < mid_idx:
        return binarySearch(arr, target, lb, mid_idx-1)
    else:
        return binarySearch(arr, target, mid_idx+1, ub)
```

# Q4
Please implement a Bubble Sort function.

Your implementation should take in 1 parameter:

- an array of objects.

Your implementation should also return an array with the contents of the given array sorted in ascending order.

```python
def bubbleSort(arr):
    for i in range(len(arr)):
        for j in range(len(arr)-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
```

# Q5
Please implement an Insertion Sort function.

Your implementation should take in 1 parameter:

- an array of objects.

Your implementation should also return an array with the contents of the given array sorted in ascending order.

```python
def insertion_sort(A):
    for i in range(len(A)):
        j = i
        while j > 0 and A[j - 1] > A[j]:
            A[j], A[j - 1] = A[j - 1], A[j]
            j = j - 1
    return A
```

# Q6
Please implement a Quicksort function.

Your implementation should take in 1 parameter:

- an array of objects.

Your implementation should also return an array with the contents of the given array sorted in ascending order.



```python
def quickSort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[0]
    left = []
    right = []
    for i in arr[1:]:
        if i < pivot:
            left.append(i)
        else:
            right.append(i)
    return quickSort(left) + [pivot] + quickSort(right)
```

# Q7
Please implement a Base-k to Denary Positive Integer Conversion function.

Your implemntation should take in 2 parameters:

- a string corresponding to a positive integer value represented in base k (e.g., if k is 2, the integer is represented in Binary), and
- an integer defining the value of k (between 2 and 16).

Your implementation should also return the integer value in Denary (i.e., base 10) - note that your return value should be of type integer.

```python
def basek_to_denary(basek_value, k):
    mapping = "0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    basek_value = str(basek_value)[::-1]
    denary = 0
    for i in range(len(basek_value)):
        denary += mapping.index(basek_value[i]) * k**i
    return denary
```

# Q8
Please implement a Denary to Base-k Positive Integer Conversion function.

Your implementation should take in 2 parameters:

- a positive integer value, and
- an integer defining the value of k (between 2 and 16), where k is the intended base for the desired representation of the given integer.

Your implementation should also return a string corresponding to a positive integer value represented in base k (e.g., if k is 2, this string represents the given integer in Binary).

```python
def denary_to_basek(denary, k):
    mapping = "0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    base_k_value = ""
    while denary > 0:
        base_k_value = mapping[denary % k] + base_k_value
        denary //= k
    return base_k_value
```

```python
class Square:
    def __init__(self, length):
        self._length = length
    
    def getLength(self):
        return self._length
    
    def computePerimeter(self):
        return float(self._length * 4)
    
    def computeArea(self):
        return float(self._length * self._length)

class Rectangle(Square):
    def __init__(self, length, breath):
        self._breath = breath
        super().__init__(length)
    
    def getBreath(self):
        return self._breath
    
    def computePerimeter(self):
        return float(self._length *2 + self._breath*2)
    
    def computeArea(self):
        return float(self._length * self._breath)
```

```python
class HashTable:

    def __init__(self, n):
        self._array = [None for _ in range(n)]
        self._size = n

    def hash(self, s):
        s = s[::-1]
        return sum([ord(char) * count+1 for (count, char) in enumerate(s)]) % self._size

    def insert(self, data):
        index = self._exists(data, True)
        if index == -1:
            return False
        else:
            self._array[index] = data

    def _exists(self, target, insert_flg=False):
        index = self.hash(str(target))
        if insert_flg == True:
            target = None
        if self._array[index] == target:
            return index
        else:
            i = (index + 1) % self._size
            while i != index:
                if self._array[i] == target:
                    return i
                else:
                    i = (i + 1) % self._size
            return -1

    def exists(self, data):
        if self._exists(data) == -1:
            return False
        else:
            return True

    def delete(self, data):
        index = self._exists(data)
        if index == -1:
            return False
        else:
            self._array[index] = None
            return True

    def __str__(self):
        return str(self._array)
```

```python
# BST
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

    def __str__(self):
        return str(self.data)


class BST:
    def __init__(self):
        self.root = None
    
    def print(self):
        self.inOrder()
    
    def insert(self, data):
        new_node = Node(data)
        if self.root == None:
            self.root = new_node
            return
        cur = self.root
        while True:
            if data < cur.data:
                # Insert left
                if cur.left == None:
                    cur.left = new_node
                    return
                else:
                    cur = cur.left
            else:
                # Insert right
                if cur.right == None:
                    cur.right = new_node
                    return
                else:
                    cur = cur.right

    def inOrder(self):
        def _inOrder(node):
            if node == None:
                return
            _inOrder(node.left)
            print(node)
            _inOrder(node.right)
        _inOrder(self.root)

    def preOrder(self):
        def _preOrder(node):
            if node == None:
                return
            print(node)
            _preOrder(node.left)
            _preOrder(node.right)
        _preOrder(self.root)

    def postOrder(self):
        def _postOrder(node):
            if node == None:
                return
            _postOrder(node.left)
            _postOrder(node.right)
            print(node)
        _postOrder(self.root)

    def count(self):
        cur = self.root
        def _count(cur):
            if cur == None:
                return 0
            return 1 + _count(cur.left) + _count(cur.right)
        return _count(cur)
    
    def exists(self, data):
        cur = self.root

        def _exists(cur, data):
            if cur.data == data:
                return True
            if data < cur.data:
                if cur.left == None:
                    return False
                return _exists(cur.left, data)
            else:
                if cur.right == None:
                    return False
                return _exists(cur.right, data)
        return _exists(cur, data)
```
